```python
emergency_status = []     #list created to store emergency patients
normal_status = []        #list created to store normal patients
bedStatus = {'Bed 1': 'Unoccupied', 'Bed 2': 'Unoccupied', 'Bed 3': 'Unoccupied'} #dictionary showing the status of beds
patientDetails = {'Name': '', 'Age': '', 'Emergency': ''}    #Dictionary to store the pateint details
flag = True                     #To run the program recursively, Flag has been maintained.


def inputData():                            #Function to take input on patient details
    patientList = {}
    print('Please enter the name of the Patient ')
    name = str(input())
    print('Please enter the age of the Patient ')
    age = int(input())
    print('Is it an Emergency ? [y/n]')
    emergencyStatus = str(input())             #If emergency status is 'y', details are stored in emergency_status
    if emergencyStatus.lower() == 'y':
        emergency_status.append(name)
    elif emergencyStatus.lower() == 'n':      #If emergency status is 'n' details stored in normal_status
        normal_status.append(name)
    else:
        print('Invalid Entry')
        inputData()
    patientList["Name"] = name
    patientList["Age"] = age
    patientList["Emergency"] = emergencyStatus
    if bedStatus['Bed 1'] == 'Unoccupied': #Condition to assign a bed and store the patient details with bed number
        patientList["BedNumber"] = 'B1'
    elif bedStatus['Bed 2'] == 'Unoccupied':
        patientList["BedNumber"] = 'B2'
    else:
        patientList["BedNumber"] = 'B3'
    return patientList


def addPatient():              #Function to add patient
    patientList = {}
    if 'Unoccupied' in bedStatus.values():  #If bedStatus is 'unoccupied', program will take input data of patients
        patientList = inputData()
        for i in bedStatus:
            if bedStatus[i] == 'Unoccupied':
                bedStatus[i] = patientList
                break
            else:
                continue
    else:                          #If all beds are occupied, a message pops up showing all beds occupied
        print('Sorry! All beds are occupied :(')



def showPatients():                         # Function to display the bed status:
    for val in bedStatus:              
        print("************************************")
        print("PATIENT DETAILS IN", val)
        print("************************************")
        if bedStatus[val] != "Unoccupied":   
            for x in bedStatus[val]:
                print(x, "->", (bedStatus[val])[x])
        else:
            print(bedStatus[val])
    print("************************************")


def releasePatient(bedNumber):         #Function to release an added patient from bedStatus
    if bedNumber == 'B1':
        bedStatus['Bed 1'] = 'Unoccupied'
    elif bedNumber == 'B2':
        bedStatus['Bed 2'] = 'Unoccupied'
    elif bedNumber == 'B3':
        bedStatus['Bed 3'] = 'Unoccupied'


def main():    #Main function that acts as the user interface and entry point to the program
    global flag
    while flag:          #As long as the flag value is true, the program runs recursively
        print('                WELCOME TO PATIENT MANAGEMENT SYSTEM')
        print('*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*')
        print('* Choose your selection(Enter the number corresponding to your selection):  *')
        print('* 1: CHECK AVAILABILITY OF BEDS                                             *')
        print('* 2: ADD PATIENT                                                            *')
        print('* 3: RELEASE PATIENT                                                        *')
        print('* 4: EXIT                                                                   *')
        print('*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*')
        user_input = int(input("Enter your Selection : ")) 
        if user_input == 1:
            showPatients()
        elif user_input == 2:
            addPatient()
        elif user_input == 3:
            bed_number = (input('Enter the Bed Number to Release: '))
            if bed_number in ['B1', 'B2', 'B3']:
                releasePatient(bed_number)
                print('Bed Released!')
            else:
                print('Bed with this number is not available')
        elif user_input == 4:
            print('THANK YOU!!')
            break
        else:
            print('INCORRECT SELECTION')
            flag = False

    user_str = input("GO BACK TO MAIN MENU?: (yes/no)")
    if user_str.lower() == 'yes':
        main()
    else:
        flag = False


main()
```
